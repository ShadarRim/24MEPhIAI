```python
import pandas as pd
df = pd.read_csv('laptop_price.csv', encoding='ISO-8859-1')
df.head()
```

```python
df.info()
```

В датасете есть типы int64, object, float56
Пропусков в датасете нет 

```python
#очистка столбца ram
df['Ram'] = df['Ram'].astype(str)
df['Ram'] = df['Ram'].str.extract(r'(\d+)')
df['Ram'] = df['Ram'].astype(int)
#очищаем Memory
df['Memory'] = df['Memory'].astype(str)
df['Memory'] = df['Memory'].str.extract(r'(\d+)')
df['Memory'] = df['Memory'].astype(int)
#разделение на ширину и длину
df[['Width', 'Height']] = df['ScreenResolution'].str.extract(r'(\d+)x(\d+)', expand = True)
df['Width'] = df['Width'].astype(int)
df['Height'] = df['Height'].astype(int)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#строим матрицу корреляции для числовых переменных в датасете 
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only = True), xticklabels=df.corr(numeric_only = True).columns, yticklabels=df.corr(numeric_only = True).columns, cmap='BuPu', center=0, annot=True)
plt.title('Матрица корреляции', fontsize=22)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
#Постройте boxplot для всех переменных и выведете объекты которые являются выбросами (строки из датафрейма).
fig, axes = plt.subplots(3, 2, figsize=(15,15))

sns.boxplot(df['Inches'], ax=axes[0, 0])
sns.boxplot(df['Ram'], ax=axes[0, 1])
sns.boxplot(df['Memory'], ax=axes[1, 0])
sns.boxplot(df['Price_euros'], ax=axes[1, 1])
sns.boxplot(df['Width'], ax=axes[2, 0])
sns.boxplot(df['Height'], ax=axes[2, 1])

print(df['Inches'].median())
print(df['Ram'].median())
print(df['Memory'].median())
print(df['Price_euros'].median())
print(df['Width'].median())
print(df['Height'].median())
```
