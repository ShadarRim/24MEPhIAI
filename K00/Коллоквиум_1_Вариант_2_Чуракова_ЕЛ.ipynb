## NumPy

Вам дан одномерный массив. На основе этого массива сгенерируйте двумерный массив, который группирует по 4 элемента в строку из исходного.

Размерность исходного массива кратна 4.

Например:

`A = [1,2,3,4,5]` - исходный массив

`B = [[1,2,3,4],[2,3,4,5]]` - полученный

```python
import numpy as np
import re
a = np.array([1,2,3,4,5,6,7, 8])
b = np.array([a[i:i+4:] for i in range(len(a)-3)])
b
```

```python
import pandas as pd
import re
df = pd.read_csv('laptop_price.csv', encoding='latin-1')
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
print(df['Ram'].unique())
df['Ram'] = pd.to_numeric(df['Ram'].str[:-2])
```

```python
df[['ProcessorType', 'Frequency']] = df['Cpu'].str.rsplit(n=1, expand = True)
print(df['Frequency'].unique())
df['Frequency'] = pd.to_numeric(df['Frequency'].str[:-3])
print(df['Weight'].unique())
df['Weight'] = pd.to_numeric(df['Weight'].str[:-2])
```

```python
df_1 = df['ScreenResolution'].str.replace('x', ' ')
df_2 = df_1.str.split(' ')
height = pd.to_numeric(df_2.str[-2])
width = pd.to_numeric(df_2.str[-1])
df['Height'] = height
df['Width'] = width
df['ScreenType'] = df['ScreenResolution'].str.extract(r'^(.*?)(?=\d+x\d+|$)', expand=False).str.strip()
```

```python
def extract_size(storage, storage_type):
    regex = rf'(\d+(\.\d+)?)(TB|GB) {storage_type}'
    matches = df['Memory'].str.extractall(regex)
    sizes = matches[0].astype(float) * np.where(matches[2] == 'TB', 1024, 1)
    return sizes.groupby(level=0).sum()


df['Flash Storage'] = extract_size(df['Memory'], 'Flash Storage').reindex(df.index, fill_value=0)
df['Hybrid'] = extract_size(df['Memory'], 'Hybrid').reindex(df.index, fill_value=0)
df['SSD'] = extract_size(df['Memory'], 'SSD').reindex(df.index, fill_value=0)
df['HDD'] = extract_size(df['Memory'], 'HDD').reindex(df.index, fill_value=0)

df.head()
```

```python
df.drop(columns=['Memory', 'Cpu', 'ScreenResolution'], inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(numeric_only = True), xticklabels=df.corr(numeric_only = True).columns, yticklabels=df.corr(numeric_only = True).columns, cmap='RdYlGn', center=0, annot=True)

plt.title('Матрица корреляции для всех переменных', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.pairplot(df, kind="reg", hue = 'Company')
plt.show()
```

```python
df_1 = df[['laptop_ID', 'Inches', 'Ram', 'Weight', 'Price_euros', 'Frequency',
       'Height', 'Width', 'Flash Storage', 'Hybrid', 'SSD', 'HDD']]
outliers = pd.DataFrame(columns=df_1.columns)
for column in df_1.columns:
        Q1 = df_1[column].quantile(0.25)
        Q3 = df_1[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        condition = (df_1[column] < lower_bound) | (df_1[column] > upper_bound)
        outliers = pd.concat([outliers, df[condition]])
        pd.set_option('display.max_columns', None)  # Показывать все столбцы
        pd.set_option('display.width', 1000)
        print(outliers)
```
