## Условная вероятность. Независимые события. Формула Байеса.

Наступление одного события может влиять на наступление другого, то есть события могут быть взаимосвязаны или чем-то "обусловлены". Например, вероятность того, что за день хоть раз выпадет снег, выше зимой.

__Условная вероятность__ $P(A|B)$ — это вероятность наступления события $A$ при условии, что произошло событие $B$.
$$P(A | B) = \dfrac{P(AB)}{P(B)}$$

События $A$ и $B$ называются __независимыми__, если $P(A|B) = P(A)$, т.е. если наступление события $B$ не влияет на вероятность наступления события $A$, **и** наоборот. Например, при многократном броске кубика результаты одного броска никак не влияют на результаты других бросков. Напротив, события «зима» и «снег» из примера выше, являются зависимыми.

_Замечание_. Зависимость событий не означает, что одно гарантированно влечёт другое. Она лишь означает, что наступление одного __меняет вероятность__ наступления другого.

Для __независимых__ событий:
$$P(AB) = P(A) \cdot P(B)$$

__Формула Байеса__ позволяет «развернуть в обратную сторону» условную вероятность $P(A|B)$, т.е. выразить её через $P(B|A)$. 

По определению условной вероятности:
$$P(A|B) = \dfrac{P(AB)}{P(B)}$$
Совместную вероятность $P(AB)$ можно теперь выразить в обратном порядке:
$P(AB) = P(B|A) \cdot P(A)$

Соответственно, формула Байеса:
$$P(A|B) = \dfrac{P(B|A) \cdot P(A)}{P(B)}$$

**Пример**

Три спортсмена участвуют в соревнованиях по стрельбе из лука. Один (мы не знаем какой) из трёх спортсменов стреляет и попадает в мишень. Каждый спортсмен стреляет один раз в равных с другими условиями. Вероятность такого события для первого спортсмена равна $0.2$, для второго — $0.4$, для третьего — $0.7$.

_Задача_: найти вероятность того, что стрелял третий спортсмен.

Определим рассматриваемые в задачи события: событие $A$ заключается в том, что произошло попадание в мишень. Кроме того, пусть для каждого $i \in \{ 1, 2, 3 \}$ событие $B_i$ заключается в том, что выстрелил $i$-й спортсмен.

В такой формулировке наша задача — найти вероятность $P(B_3 | A)$: то есть найсти вероятность того, что попал третий спортсмен при условии, что попадание было вообще. 

По формуле Байеса:
$$P(B_3|A) = \frac{P(A|B_3) \cdot P(B_3)}{P(A)}.$$

Ввероятность каждого события $B_i$ равна $P(B_i) =  1/3$ (это дано в условии).

Кроме того, нам даны вероятности попадания в мишень для каждого из спортсменов:
$$P(A|B_1) = 0.2$$
$$P(A|B_2) = 0.4$$
$$P(A|B_3) = 0.7$$

Остаётся лишь оценить безусловную вероятность попадания в мишень — $P(A)$. Для этого воспользуемся формулой полной вероятности:
$$P(A) = P(A | B_1) \cdot P(B_1) + P(A | B_2) \cdot P(B_2) + P(A | B_3) \cdot P(B_3).$$

Имеем:
$$P(A) = 0.2 \cdot \dfrac{1}{3} + 0.4 \cdot \dfrac{1}{3} + 0.7 \cdot \dfrac{1}{3} = \dfrac{13}{30}$$

Итак, искомая вероятность:
$$P(B_3|A) = \dfrac{0.7 \cdot \frac{1}{3}}{\frac{13}{30}} = \frac{7}{13} \approx 0.5385$$
